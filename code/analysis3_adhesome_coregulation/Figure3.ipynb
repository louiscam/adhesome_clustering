### Assess difference between the two mutlivariate distributions

```python
# Permutation test for two sample mutivariate goodness of fit
# See: https://projecteuclid.org/journals/annals-of-statistics/volume-16/issue-2/
# A-Multivariate-Two-Sample-Test-Based-on-the-Number-of/10.1214/aos/1176350835.full

# 0 - Data
x1 = np.array(filtered_combined_edgelist['scaled_hic'])
y1 = np.array(filtered_combined_edgelist['corr'])
x2 = np.array(filtered_rand_combined_edgelist['scaled_hic'])
y2 = np.array(filtered_rand_combined_edgelist['corr'])

# 1- Format the data
Z0 = np.stack([x1,y1], axis=1) # sample from distribution 0
Z1 = np.stack([x2,y2], axis=1) # sample from distribution 1
n0 = len(x1) # sample size from distribution 1
n1 = len(x2) # sample size from distribution 2
n = n0+n1

# 2- Pool together the data
Z = np.concatenate([Z0,Z1], axis=0)

# 3- Obtain k-NN network
k = 1
A = kneighbors_graph(Z, n_neighbors=k, mode='connectivity', metric='minkowski', p=2, include_self=False)
A = pd.DataFrame(A.toarray())

# 4- Count number of coincidences
T = (A.iloc[:n0,:n0].sum().sum()+A.iloc[n0:n,n0:n].sum().sum())/n
```

```python
# 6- Plot distribution and compute p-value
with open(saving_dir+'null_stats.pkl', 'rb') as f:
    null_stats0 = pickle.load(f)
null_stats = np.concatenate([null_stats0, [T]])
pval = np.mean(null_stats>=T)

sns.set(font_scale=1.2)
plt.figure()
sns.histplot(null_stats0, stat='density', bins=np.arange(0.49, 0.540, 0.002))
plt.vlines(x=T, ymin=0, ymax=140, color='red', linestyle='dashed', alpha=0.5,
           label=f'p-value = {np.round(pval,4)}')
plt.xlabel('Proportion of class coincidences')
plt.ylabel('Density')
plt.legend(loc='upper right')
# plt.savefig(saving_supp_dir+f'suppfig_4f.pdf', format='pdf')
plt.show()
print('Test statistic = '+str(T))
print('p-value = '+str(pval))
```
