# Adhesome interactions

```python
# Load interactions
interactions_filename = 'interactions.csv'
df_interactions = pd.read_csv(dir_adhesome+interactions_filename, sep = ',', header = 0)
print('Number of interactions: '+str(len(df_interactions)))
df_interactions.head()
```

```python
# Harmonize types
df_interactions['Type'] = df_interactions['Type'].str.lower().str.replace('_reaction','')
```

```python
# Remove duplicates based on Source, Target, Type
df_interactions = df_interactions.sort_values(by='data_source')
df_interactions = df_interactions.drop_duplicates(subset=['Source','Target','Type'], keep='first')
df_interactions = df_interactions.drop_duplicates(subset=['PMID'], keep='first')
df_interactions = df_interactions.drop_duplicates(subset=['Source','Target'], keep='first')
print('Number of interactions: '+str(len(df_interactions)))
```

# Construct adhesome edge network

```python
# Construct edge list of the edge graph
rows_list = []
for i in tqdm(np.arange(len(df_interactions))):
    time.sleep(.01)
    for j in np.arange(i+1,len(df_interactions)):
        vertices1 = set([df_interactions.iloc[i].Source, df_interactions.iloc[i].Target])
        vertices2 = set([df_interactions.iloc[j].Source, df_interactions.iloc[j].Target])
        if len(vertices1.intersection(vertices2))==1:
            rows_list.append({'edge1':df_interactions.iloc[i].PMID,
                              'edge2':df_interactions.iloc[j].PMID,
                              'common_node': list(vertices1.intersection(vertices2))[0]})
pickle.dump(rows_list, open('rows_list.pkl', 'wb'))
```

```python
# Build edge list as dataframe
eg_edgelist = pd.DataFrame(rows_list)
eg_edgelist.head()
```

```python
# Map each interaction type to an integer
typeid = {'activation':0, 'binding':1, 'colocalization':2, 'compound':3,
       'dephosphorylation':4, 'direct_interaction':5, 'enzymatic':6, 'gef':7,
       'inhibition':8, 'phosphorylation':9, 'unknown':10}
# Create edge graph node metadata
eg_nodes_dict = {}
for i in tqdm(np.arange(len(df_interactions))):
    time.sleep(.01)
    eg_nodes_dict[df_interactions.iloc[i].PMID] = {'Type': df_interactions.iloc[i].Type,
                                                   'TypeID': typeid[df_interactions.iloc[i].Type]}
```

```python
# Create networkX edge network without node attributes
edge_network = nx.from_pandas_edgelist(eg_edgelist, 'edge1', 'edge2')
print('Number of nodes = '+str(len(edge_network.nodes)))
```

```python
# Add node attributes to networkx graph
nx.set_node_attributes(edge_network, eg_nodes_dict)
```

# Identify enrichment for activation/inhibition edges

```python
# Number of communities
n_comm = len(set(partition_adhesome.values()))
```

```python
# Store all subgraphs
subgraph_ls = []
for comm in range(n_comm):
    comm_nodes = [prot for prot in partition_adhesome.keys() if partition_adhesome[prot]==comm]
    subgraph = adhesome_network.subgraph(comm_nodes)
    subgraph_ls.append(subgraph)
```

```python
# Number of nodes
nnodes_ls = []
for comm in range(n_comm):
    nnodes_ls.append(len(subgraph_ls[comm].nodes))
```

```python
# Type of edges
nedges_ls, nact_ls, ninhib_ls, nplus_ls, nminus_ls = [], [], [], [], []
for comm in range(n_comm):
    comm_nodes = list(subgraph_ls[comm].nodes)
    df_comm = df_interactions[df_interactions['Source'].isin(comm_nodes) & df_interactions['Target'].isin(comm_nodes)]
    nedges_ls.append(len(df_comm))
    nact_ls.append(len(df_comm[df_comm['Type']=='activation']))
    ninhib_ls.append(len(df_comm[df_comm['Type']=='inhibition']))
    nplus_ls.append(len(df_comm[df_comm['Effect']=='+']))
    nminus_ls.append(len(df_comm[df_comm['Effect']=='_']))
```

```python
# Create summary dataframe
network_summary = pd.DataFrame({'cluster': range(n_comm),
                                'size': nnodes_ls,
                                'edges': nedges_ls,
                                'activation': nact_ls,
                                'inhibition': ninhib_ls,
                                '+': nplus_ls,
                                '_': nminus_ls})
network_summary
```
